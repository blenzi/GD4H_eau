```python
import pandas as pd
import geopandas as gpd
from IPython.display import Markdown as md
```

```python
data_AtlaSante = pd.read_csv('data/info_AtlaSante.csv').set_index('Région')
data_AtlaSante
```

```python
cartes_AtlaSante = {region: gpd.read_file(data['Fichier']) for region, data in data_AtlaSante.iterrows()}
```

```python
for region, df in cartes_AtlaSante.items():
    annee = data_AtlaSante.loc[region, 'Année']
    unique_entries = df.drop_duplicates(data_AtlaSante.loc[region, 'Champ AtlaSante']).shape[0]
    no_geometry = df[df['geometry'].isna()]
    display(md(f"### {region}, entries: {df.shape[0]}, unique: {unique_entries}, no geometry: {no_geometry.shape[0]}"))
    display(md(f"### Duplicates:"))
    
    df_temp = df.set_index(data_AtlaSante.loc[region, 'Champ AtlaSante']).sort_index()
    display(df_temp.loc[df_temp.index.duplicated(False)])
    df_temp.loc[df_temp.index.duplicated(False)].to_csv(f"outputs/{region}_{annee}_duplicates.csv")
    no_geometry.to_csv(f"outputs/{region}_{annee}_noGeometry.csv")
```
